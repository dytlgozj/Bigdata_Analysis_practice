```python
# 2
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

train = pd.read_csv("3rd_TravelInsurancePrediction_train.csv")
test = pd.read_csv("3rd_TravelInsurancePrediction_test.csv")

# print(train.shape)
# print(test.shape)
# print(train.describe())
# print(test.describe())

x_train = train.drop(["TravelInsurance", "ID"], axis = 1)
x_train.head()
y = train[["TravelInsurance"]]
# y.head()
x_test = test.drop("ID", axis = 1)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

x_train_cat = x_train[["Employment Type", "GraduateOrNot", "ChronicDiseases", "FrequentFlyer", "EverTravelledAbroad"]]
x_train_not_cat = x_train[["Age", "AnnualIncome", "FamilyMembers"]]

x_test_cat = x_test[["Employment Type", "GraduateOrNot", "ChronicDiseases", "FrequentFlyer", "EverTravelledAbroad"]]
x_test_not_cat = x_test[["Age", "AnnualIncome", "FamilyMembers"]]

x_train_cat = pd.get_dummies(x_train_cat)
x_test_cat = pd.get_dummies(x_test_cat)
x_train_cat, x_test_cat = x_train_cat.align(x_test_cat, join = "inner", axis = 1)

scaler.fit(x_train_not_cat)
x_scaled_train_not_cat = scaler.transform(x_train_not_cat)
x_scaled_test_not_cat = scaler.transform(x_test_not_cat)

final_train = pd.concat([x_train_cat, pd.DataFrame(x_scaled_train_not_cat)], axis = 1)
final_test = pd.concat([x_test_cat, pd.DataFrame(x_scaled_test_not_cat)], axis = 1)

# final_train.head()

from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(final_train, y, random_state = 42, stratify = y, test_size = 0.2)

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state = 42)
model.fit(X_train, y_train)
print(f"train score : {model.score(X_train, y_train)}")
print(f"validation score : {model.score(X_valid, y_valid)}")

prob_valid = model.predict_proba(X_valid)
# print(prob_valid)
from sklearn.metrics import roc_auc_score
# help(roc_auc_score)
print(roc_auc_score(y_valid, prob_valid[:, 1]))

prob_test = model.predict_proba(final_test)
# print(prob_test)
final_ans = pd.DataFrame({"ID": test["ID"], "pred": prob_test[:, 1]})
# final_ans.head(10)
final_ans.to_csv("2016003790.csv", index = False)
```
