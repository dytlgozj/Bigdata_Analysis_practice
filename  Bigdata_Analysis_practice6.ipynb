```python
# 2
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings("ignore")

train = pd.read_csv("X_train.csv")
test = pd.read_csv("X_test.csv")
y = pd.read_csv("y_train.csv")
scaler = MinMaxScaler()

# print(train.head())
# print(test.head())
# print(y.head())
# print(train.shape)
# print(test.shape)
# print(y.shape)

x_train = train.drop("ID", axis = 1)
y = y[["Reached.on.Time_Y.N"]]
x_test = test.drop("ID", axis = 1)

# x_train.head()

def _func(series):
    if series == "high":
        return 3
    elif series == "medium":
        return 2
    elif series == "low":
        return 1

x_train["Product_importance"] = x_train["Product_importance"].apply(_func)
x_train_cat = x_train[["Warehouse_block", "Mode_of_Shipment", "Gender"]]
x_train_not_cat = x_train[["Customer_care_calls", "Customer_rating", "Cost_of_the_Product", "Prior_purchases", "Product_importance", "Discount_offered", "Weight_in_gms"]]

x_test["Product_importance"] = x_test["Product_importance"].apply(_func)
x_test_cat = x_test[["Warehouse_block", "Mode_of_Shipment", "Gender"]]
x_test_not_cat = x_test[["Customer_care_calls", "Customer_rating", "Cost_of_the_Product", "Prior_purchases", "Product_importance", "Discount_offered", "Weight_in_gms"]]

x_train_cat = pd.get_dummies(x_train_cat)
x_test_cat = pd.get_dummies(x_test_cat)

x_train_cat, x_test_cat = x_train_cat.align(x_test_cat, join = "inner", axis = 1)

scaler.fit(x_train_not_cat)
x_train_not_cat = scaler.transform(x_train_not_cat)
x_test_not_cat = scaler.transform(x_test_not_cat)

final_train = pd.concat([x_train_cat, pd.DataFrame(x_train_not_cat)], axis = 1)
final_test = pd.concat([x_test_cat, pd.DataFrame(x_test_not_cat)], axis = 1)

# final_train.head(7)

from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(final_train, y, random_state = 42, stratify = y, test_size = 0.2)

# help(RandomForestClassifier)


from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state = 42)
model.fit(X_train, y_train)
pred_valid = model.predict_proba(X_valid)
print(f"train score : {model.score(X_train, y_train)}")
print(f"validation score : {model.score(X_valid, y_valid)}")

from sklearn.metrics import roc_auc_score
print(f"roc_auc_score : {roc_auc_score(y_valid, pred_valid[:, 0])}")

pred_test = model.predict_proba(final_test)
final_ans = pd.DataFrame({"ID": test["ID"], "predict_prob": pred_test[:, 0]})
final_ans.to_csv("2016003790.csv", index = False)
```
